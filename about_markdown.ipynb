# ------------------- Raw text -------------------


**Syntax**

Today is a cloudy day and a good day to practice Markdown.

&darr;

**Output**

Today is a cloudy day and a good day to practice Markdown.

___
**Syntax**

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Today is a cloudy day and a good day to practice Markdown.

&darr;

**Output (Code Blocks)** 

    Today is a cloudy day and a good day to practice Markdown.

___
**Syntax**

\>Today is a cloudy day and a good day to practice Markdown.

&darr;

**Output (Quote)**

>Today is a cloudy day and a good day to practice Markdown.

# ------------------- HTML tag -------------------

It respresents the root of an HTML document.

**Syntax**

\<div>

&darr;

**Output**
    
<div>

___

**Syntax**

\\\<div>

&darr;

**Output**
    
\<div>

___

**Syntax**

\&lt;

&darr;

**Output**
    
&lt;

___

**Syntax**

\&lt;div>

&darr;

**Output**
    
&lt;div>

___

**Syntax**

\\\\\<div>

&darr;

**Output**
    
\\<div>

___

**Syntax**

<div\\>

or 

\\<div\\>

&darr;

**Output**
    
<div\>
